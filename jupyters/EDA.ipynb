## 按总票房金额将电影进行分类
#### 第一步，汇总统计不同电影的总票房

```python
datadir = '../data/2019-2024电影每日票房信息'
movie_data_filename_list = glob.glob(os.path.join(datadir, '*.xlsx'))
total_box_office_map = {}
for item in movie_data_filename_list:
    movie_name = item.split('\\')[1].strip('.xlsx')
    item_data = pd.read_excel(item, index_col=0)
    total_box_office = item_data['当前票房(万)'].sum() / 10000
    total_box_office_map[movie_name] = total_box_office
with open('../data/电影总票房统计表.csv', 'w', encoding='utf-8') as f:
    for key, value in total_box_office_map.items():
        f.write(f'{key},{value:.2f}\n')
```

```python
figure = plt.figure(figsize=(12, 8))
ax = figure.add_subplot(111)
tmp = np.array([i for i in total_box_office_map.values()]) + 1.0
_ = plt.hist(np.log(tmp), bins=30, color='k', alpha=0.6)
```

上图的横轴是经过对数变换的。选择分组的区间为[0,4),[4,8),[8,15]。

```python
#将电影人为地分成了三个等级，如：普通电影、热门电影、爆款电影

group_1_movie_list = []
group_2_movie_list = []
group_3_movie_list = []

for key, value in total_box_office_map.items():
    if value < np.exp(4):
        group_1_movie_list.append(key)
    elif value < np.exp(8):
        group_2_movie_list.append(key)
    else:
        group_3_movie_list.append(key)
```

```python
def gene_avg_box_office_seq(group_movie_list, target_column='场均收入'):
    box_office_seq_list = []
    count_seq_list = []
    for movie_name in group_movie_list:
        movie_path = os.path.join(datadir, f'{movie_name}.xlsx')
        item_data = pd.read_excel(movie_path, index_col=0)
        box_office_seq = list(item_data[target_column].values / 10000.0)
        box_office_seq = box_office_seq + [0] * (60 - len(box_office_seq))
        count_seq = [1] * len(box_office_seq) + [0] * (60 - len(box_office_seq))
        box_office_seq_list.append(box_office_seq)
        count_seq_list.append(count_seq)
    box_office_seq_array = np.array(box_office_seq_list)
    count_seq_array = np.array(count_seq_list)
    return np.sum(box_office_seq_array, axis=0) / np.sum(count_seq_array, axis=0)

group_1_bfs = gene_avg_box_office_seq(group_1_movie_list)
group_2_bfs = gene_avg_box_office_seq(group_2_movie_list)
group_3_bfs = gene_avg_box_office_seq(group_3_movie_list)
```

```python
def gene_movie_calendar() -> pd.DataFrame:
    if os.path.exists('../data/movie_calendar.csv'):
        # 读取该文件
        movie_online_date_df = pd.read_csv('../data/movie_calendar.csv')
        movie_online_date_df['上映日期'] = pd.to_datetime(movie_online_date_df['上映日期'])
        movie_online_date_df['下线日期'] = pd.to_datetime(movie_online_date_df['下线日期'])
        return movie_online_date_df
    else:
        movie_filename_list = glob.glob(os.path.join('../data/2019-2024电影每日票房信息', '*.xlsx'))
        movie_online_date_list = []
        for item in movie_filename_list:
            movie_name = item.split('\\')[1].strip('.xlsx')
            item_data = pd.read_excel(item, index_col=0)
            online_date = item_data['日期/上映天数'].iloc[0].split('|')[0]
            offline_date = item_data['日期/上映天数'].iloc[-1].split('|')[0]
            movie_online_date_list.append([movie_name, online_date, offline_date])
        movie_online_date_df = pd.DataFrame(movie_online_date_list, columns=['电影名称', '上映日期', '下线日期'])
        # 按照上映日期升序排列
        movie_online_date_df['上映日期'] = pd.to_datetime(movie_online_date_df['上映日期'])
        movie_online_date_df['下线日期'] = pd.to_datetime(movie_online_date_df['下线日期'])
        movie_online_date_df.sort_values(by='上映日期', inplace=True)
        # 写入到硬盘
        movie_online_date_df.to_csv('../data/movie_calendar.csv', index=False)
        return movie_online_date_df

movie_calendar = gene_movie_calendar()
current_date = (pd.to_datetime('2019-03-01') +
                             pd.to_timedelta(np.random.randint(0, 365 * 5 + 120), unit='D'))
valid_index = (movie_calendar['上映日期'] < current_date) & (movie_calendar['下线日期'] > current_date)
valid_data = movie_calendar[valid_index]
valid_data
```

```python
import numpy as np
import matplotlib.pyplot as plt


def custom_function(x):
    if isinstance(x, (int, float)):
        if x < 25000:
            return x
        else:
            a = 21000
            d = -190000
            return a * np.log(x) + d
    else:
        y = np.zeros_like(x)
        mask1 = x < 25000
        mask2 = x >= 25000
        y[mask1] = x[mask1]
        a = 21276.59574468
        d = -190440.40425532
        y[mask2] = a * np.log(x[mask2]) + d
        return y


# 生成 x 值
x = np.linspace(0, 50000, 1000)
y = custom_function(x)

# 绘制曲线
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Custom Monotonic Increasing Function')
plt.plot([0, 50000], [0, 50000], lw=2.0, alpha=0.5, color='r', ls='--')
plt.grid(True)
plt.show()
```

```python
# 1. 导入必要的库
import pandas as pd
import numpy as np
import glob
import os
from tqdm import tqdm # 用于显示进度条

# 2. 定义和环境(SimpleScheduler)中完全一致的参数
# 确保这里的列名与您环境代码中的 feature_columns 完全一致
feature_columns = ['当前票房(万)', '当前场次', '当前人次(万)', '票房占比', '场次占比', '人次占比',
                   '当前平均票价', '场均人次', '场均收入', '周索引']

datadir = '../data/2019-2024电影每日票房信息'

# 3. 初始化用于存储全局最大/最小值的变量
# 我们有9个经营指标 + 7个独热编码的星期特征 = 16个最终特征
num_features = 16
global_min_vals = np.full(num_features, np.inf)  # 初始化为正无穷大
global_max_vals = np.full(num_features, -np.inf) # 初始化为负无穷大

# 4. 遍历所有Excel文件
movie_data_filename_list = glob.glob(os.path.join(datadir, '*.xlsx'))
print(f"开始扫描 {len(movie_data_filename_list)} 个电影数据文件...")

for movie_filename in tqdm(movie_data_filename_list):
    try:
        # --- 这部分代码完全模拟了您环境中 reset 方法的数据处理流程 ---
        
        # a. 加载数据
        item_data = pd.read_excel(movie_filename, index_col=0)
        
        # b. 处理“日期/上映天数”列，得到“周索引”
        week_index_map = {'一': 0, '二': 1, '三': 2, '四': 3, '五': 4, '六': 5, '日': 6}
        week_index_list = []
        # 检查列是否存在，避免因数据文件格式问题报错
        if '日期/上映天数' not in item_data.columns:
            print(f"警告：文件 {movie_filename} 缺少 '日期/上映天数' 列，已跳过。")
            continue
            
        for item in item_data['日期/上映天数']:
            try:
                week_index_list.append(week_index_map[item.split('|')[2]])
            except (IndexError, KeyError):
                week_index_list.append(0) # 如果格式错误，默认填充周一
        item_data['周索引'] = week_index_list
        
        # c. 筛选出我们需要的10个原始特征
        # 检查所有需要的列是否存在
        if not all(col in item_data.columns for col in feature_columns):
            print(f"警告：文件 {movie_filename} 缺少必要的特征列，已跳过。")
            continue
            
        item_data_features = item_data[feature_columns]

        # d. 对“周索引”进行独热编码
        numeric_features = item_data_features.iloc[:, :-1].values # 取出前9个经营指标
        week_indices = item_data_features.iloc[:, -1].values   # 取出周索引列
        
        # 创建一个单位矩阵，利用索引快速完成独热编码
        one_hot_weeks = np.eye(7)[week_indices]
        
        # e. 拼接成最终的16个特征
        processed_features = np.concatenate([numeric_features, one_hot_weeks], axis=1)

        # --- 数据扫描核心步骤 ---
        
        # f. 计算当前文件的最大/最小值
        # axis=0 表示沿着行的方向计算，得到每一列的最大/最小值
        current_min = np.min(processed_features, axis=0)
        current_max = np.max(processed_features, axis=0)
        
        # g. 更新全局最大/最小值
        global_min_vals = np.minimum(global_min_vals, current_min)
        global_max_vals = np.maximum(global_max_vals, current_max)

    except Exception as e:
        print(f"处理文件 {movie_filename} 时发生错误: {e}")
        continue

# 5. 打印最终结果
print("\n--- 数据扫描完成！---")
print("请将以下计算出的最大最小值，复制到 C51Agent 的 feature_engineer 方法中。\n")

# 为了方便复制，格式化输出
min_vals_str = ", ".join([f"{val:.2f}" for val in global_min_vals])
max_vals_str = ", ".join([f"{val:.2f}" for val in global_max_vals])

print("min_vals = tf.constant([", min_vals_str, "], dtype=tf.float32)")
print("max_vals = tf.constant([", max_vals_str, "], dtype=tf.float32)")

# 为了方便核对，我们把每个值对应的特征名也打印出来
final_feature_names = feature_columns[:-1] + ['周一', '周二', '周三', '周四', '周五', '周六', '周日']
results_df = pd.DataFrame({
    '特征名称': final_feature_names,
    '最小值': global_min_vals,
    '最大值': global_max_vals
})
print("\n--- 每个特征的具体范围如下 ---")
print(results_df.to_string())
```

## 每日在映电影数量分析
为了更好地理解市场环境，我们对数据集中每一天有多少部电影同时在映进行了统计。这可以帮助我们评估市场竞争的动态变化。

```python
import pandas as pd
import matplotlib.pyplot as plt
from tqdm import tqdm
import os

# --- 设置路径 ---
calendar_csv_path = '../data/movie_calendar.csv'

# --- 加载日历文件 ---
try:
    movie_calendar = pd.read_csv(calendar_csv_path)
    movie_calendar['上映日期'] = pd.to_datetime(movie_calendar['上映日期'])
    movie_calendar['下线日期'] = pd.to_datetime(movie_calendar['下线日期'])
    print(f"电影日历文件 '{calendar_csv_path}' 加载成功！")

except FileNotFoundError:
    print(f"错误：在路径 '{calendar_csv_path}' 下找不到 'movie_calendar.csv' 文件。")
    movie_calendar = pd.DataFrame()

# --- 计算每日电影数量 ---
if not movie_calendar.empty:
    start_date = movie_calendar['上映日期'].min()
    end_date = movie_calendar['下线日期'].max()
    date_range = pd.date_range(start=start_date, end=end_date)
    daily_movie_counts = {}

    print(f"\n正在根据日历计算从 {start_date.date()} 到 {end_date.date()} 的每日电影数量...")
    for day in tqdm(date_range):
        on_show_movies = movie_calendar[
            (movie_calendar['上映日期'] <= day) & (movie_calendar['下线日期'] >= day)
        ]
        daily_movie_counts[day] = len(on_show_movies)

    daily_counts_series = pd.Series(daily_movie_counts)

    print("\n统计完成！")
    print(f"数据覆盖的总天数: {len(daily_counts_series)} 天")
    print(f"平均每日在映电影数量: {daily_counts_series.mean():.2f} 部")
    print(f"最多一日有 {daily_counts_series.max()} 部电影同时上映")

    # --- 核心修正：在绘图前设置中文字体 ---
    plt.rcParams['font.sans-serif'] = ['SimHei']  # 指定默认字体为黑体
    plt.rcParams['axes.unicode_minus'] = False  # 解决保存图像是负号'-'显示为方块的问题
    # ------------------------------------

    # --- 绘制图表 ---
    figure = plt.figure(figsize=(15, 7))
    ax = figure.add_subplot(111)

    daily_counts_series.plot(ax=ax, color='k', alpha=0.7, label='每日在映电影数')
    daily_counts_series.rolling(window=30).mean().plot(ax=ax, color='red', linestyle='--', label='30日移动平均')

    ax.set_title('每日在映电影数量的时间趋势分析', fontsize=16)
    ax.set_xlabel('日期', fontsize=12)
    ax.set_ylabel('在映电影数量（部）', fontsize=12)
    ax.legend()
    plt.grid(True, linestyle='--', alpha=0.6)
    plt.show()
```
