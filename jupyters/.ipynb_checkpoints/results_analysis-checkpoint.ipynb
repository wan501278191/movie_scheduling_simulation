```python
import sys
import os
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib

# --- [ 步骤 0: 环境与显示设置 ] ---
# 设置matplotlib以正确显示中文
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False
# 将项目根目录添加到Python路径
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)


# --- [ 步骤 1: 数据加载与预处理 ] ---
print("--- 1. 正在加载并预处理评估数据 ---")
csv_path = '../logs/full_evaluation_results.csv'
try:
    df = pd.read_csv(csv_path)
    # 将保存为JSON字符串的列表数据，还原为真实的Python列表
    df['daily_actions'] = df['daily_actions'].apply(json.loads)
    df['daily_incomes'] = df['daily_incomes'].apply(json.loads)
    print(f"✅ 数据加载成功，共 {len(df)} 条模拟记录。")
except FileNotFoundError:
    print(f"❌ 错误：在路径 '{csv_path}' 下找不到结果文件。")
    print("👉 请先确保您已经成功运行了最新的评估脚本 main.py。")
    df = pd.DataFrame()


# --- [ 步驟 2: 定义所有分析函数 ] ---
def calculate_sharpe(daily_rewards):
    """计算夏普比率"""
    if np.std(daily_rewards) == 0:
        return 0
    return np.mean(daily_rewards) / np.std(daily_rewards)

def calculate_capture_speed(daily_actions, movie_index, threshold=0.20):
    """计算“黑马”捕获速度（天数）"""
    if movie_index is None or np.isnan(movie_index): return np.nan
    for day, schedule in enumerate(daily_actions):
        if len(schedule) > movie_index and schedule[int(movie_index)] >= threshold:
            return day + 1
    return len(daily_actions)

def calculate_cut_loss_speed(daily_actions, movie_index, threshold=0.05):
    """计算“高开低走影片”止损速度（天数）"""
    if movie_index is None or np.isnan(movie_index): return np.nan
    for day, schedule in enumerate(daily_actions):
        if len(schedule) > movie_index and schedule[int(movie_index)] <= threshold:
            return day + 1
    return len(daily_actions)


# --- [ 步骤 3: 计算并展示高级性能与风险指标 ] ---
if not df.empty:
    print("\n" + "="*80)
    print("### 核心性能与风险稳定性指标分析 ###")
    print("="*80)

    # 计算平均总收入
    summary_avg = df.groupby(['scenario', 'policy'])['total_income'].mean().unstack()
    
    # 计算性能提升百分比
    print("\n--- 核心性能指标：性能提升百分比 ---\n")
    for scenario in summary_avg.index:
        sac_income = summary_avg.loc[scenario, 'RL (SAC) 无约束']
        baselines_incomes = summary_avg.loc[scenario].drop(['RL (SAC) 无约束', 'RL (SAC) 带1%约束'])
        best_baseline_income = baselines_incomes.max()
        best_baseline_name = baselines_incomes.idxmax()
        improvement_pct = (sac_income / best_baseline_income - 1) * 100
        print(f"在【{scenario}】场景下:")
        print(f"  - SAC 无约束平均收入: {sac_income:.2f} 万元")
        print(f"  - 最佳基准 ({best_baseline_name}): {best_baseline_income:.2f} 万元")
        print(f"  - SAC 性能提升: {improvement_pct:.2f}%\n")

    # 计算最差情况表现
    summary_min = df.groupby(['scenario', 'policy'])['total_income'].min().unstack()
    print("\n--- 风险指标：最差情况表现 (万元) ---\n数值越大，抗风险能力越强：")
    print(summary_min.round(2))

    # 计算夏普比率
    df['sharpe_ratio'] = df['daily_incomes'].apply(calculate_sharpe)
    summary_sharpe = df.groupby(['scenario', 'policy'])['sharpe_ratio'].mean().unstack()
    print("\n--- 风险指标：平均夏普比率 ---\n数值越大，风险调整后收益越高：")
    print(summary_sharpe.round(2))


# --- [ 步骤 4: 自动化决策行为诊断分析 ] ---
if not df.empty:
    # 自动从数据中发现所有案例
    case_study_runs = df.dropna(subset=['case_study_movie'])
    unique_cases = case_study_runs.drop_duplicates(subset=['start_date', 'case_study_movie'])

    plots_dir = '../plots'
    os.makedirs(plots_dir, exist_ok=True)
    all_speed_results = []

    print("\n" + "="*80)
    print("### 决策行为诊断指标分析 ###")
    print("="*80)
    
    # 循环执行所有发现的案例
    for index, case_info in unique_cases.iterrows():
        case_name = case_info['case_study_movie']
        case_date = case_info['start_date']
        
        # 自动判断是黑马还是高开低走影片，并设置相应参数
        if case_name in ['流浪地球', '这个杀手不太冷静', '熊出没·伴我“熊芯”']:
            case_type = '黑马'; metric_func = calculate_capture_speed; threshold = 0.20; unit = '捕获速度 (天)'
        else:
            case_type = '高开低走'; metric_func = calculate_cut_loss_speed; threshold = 0.05; unit = '止损速度 (天)'

        print(f"\n--- 案例分析：{case_type} - 《{case_name}》 ---")
        
        case_df = df[df['start_date'] == case_date].copy()
        
        # 自动应用分析函数 (使用CSV中保存的索引)
        metric_col_name = unit.split(' ')[0]
        case_df[metric_col_name] = case_df.apply(
            lambda row: metric_func(row['daily_actions'], row['case_study_index'], threshold), axis=1
        )
        speed_summary = case_df.groupby('policy')[metric_col_name].mean().sort_values()
        speed_summary.name = f"《{case_name}》 {unit}"
        all_speed_results.append(speed_summary)
        
        print(f"速度对比 (数值越小越好):")
        print(speed_summary)
        
        # 自动绘制图表
        plt.figure(figsize=(12, 7))
        for policy_name in speed_summary.index:
            run_data = case_df[case_df['policy'] == policy_name].iloc[0]
            movie_idx = run_data['case_study_index']
            if not np.isnan(movie_idx):
                movie_idx = int(movie_idx)
                if len(run_data['daily_actions'][0]) > movie_idx:
                    schedule_over_time = [s[movie_idx] * 100 for s in run_data['daily_actions']]
                    plt.plot(range(1, len(schedule_over_time) + 1), schedule_over_time, marker='.', linestyle='-', label=policy_name)
        
        plt.axhline(y=threshold * 100, color='r', linestyle='--', label=f"阈值 ({threshold*100:.0f}%)")
        plt.title(f"各策略对“{case_type}”《{case_name}》的排片变化曲线", fontsize=16)
        plt.xlabel('模拟天数 (Day)', fontsize=12); plt.ylabel('排片占比 (%)', fontsize=12)
        plt.legend(); plt.grid(True, which='both', linestyle='--', linewidth=0.5)
        
        file_name = f"{case_name}_{metric_col_name}.png"
        save_path = os.path.join(plots_dir, file_name)
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"✅ 图表已保存至: {save_path}")
        plt.show()

    # 生成最终汇总表
    if all_speed_results:
        final_summary_df = pd.concat(all_speed_results, axis=1)
        print("\n" + "="*80)
        print("### 所有案例诊断指标汇总表 ###")
        print("数值代表天数，越小表示策略反应越快、越优秀")
        print(final_summary_df.round(2))
        print("="*80)
```

```python
import sys
import os
import pandas as pd
import numpy as np
import json
import matplotlib.pyplot as plt
import matplotlib

# --- [ 步骤 0: 环境与显示设置 ] ---
matplotlib.rcParams['font.sans-serif'] = ['SimHei']
matplotlib.rcParams['axes.unicode_minus'] = False
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
if project_root not in sys.path:
    sys.path.append(project_root)


# --- [ 步骤 1: 数据加载与预处理 ] ---
print("\n--- 1. 正在加载并预处理评估数据 ---")
csv_path = '../logs/full_evaluation_results.csv'
try:
    df = pd.read_csv(csv_path)
    df['daily_actions'] = df['daily_actions'].apply(json.loads)
    df['daily_incomes'] = df['daily_incomes'].apply(json.loads)
    print(f"✅ 数据加载成功，共 {len(df)} 条模拟记录。")
except FileNotFoundError:
    print(f"❌ 错误：在路径 '{csv_path}' 下找不到结果文件。")
    print("👉 请先确保您已经成功运行了最新的评估脚本 main.py。")
    df = pd.DataFrame()


# --- [ 步驟 2: 定义所有分析函数 ] ---
def calculate_sharpe(daily_rewards):
    """计算夏普比率"""
    if np.std(daily_rewards) == 0:
        return 0
    return np.mean(daily_rewards) / np.std(daily_rewards)

def calculate_capture_speed(daily_actions, movie_index, threshold=0.20):
    """计算“黑马”捕获速度（天数）"""
    if movie_index is None or np.isnan(movie_index): return np.nan
    for day, schedule in enumerate(daily_actions):
        if len(schedule) > movie_index and schedule[int(movie_index)] >= threshold:
            return day + 1
    return len(daily_actions)

def calculate_cut_loss_speed(daily_actions, movie_index, threshold=0.05):
    """计算“高开低走影片”止损速度（天数）"""
    if movie_index is None or np.isnan(movie_index): return np.nan
    for day, schedule in enumerate(daily_actions):
        if len(schedule) > movie_index and schedule[int(movie_index)] <= threshold:
            return day + 1
    return len(daily_actions)


# --- [ 步骤 3: 计算并展示高级性能与风险指标 ] ---
if not df.empty:
    print("\n" + "="*80)
    print("### 核心性能与风险稳定性指标分析 ###")
    print("="*80)

    summary_avg = df.groupby(['scenario', 'policy'])['total_income'].mean().unstack()
    
    print("\n--- 核心性能指标：性能提升百分比 ---\n")
    for scenario in summary_avg.index:
        
        # ========================= [ 关键修改：开始 ] =========================
        # 主模型依然是“无约束”版本
        sac_policy_name = 'RL (SAC) 无约束'
        sac_income = summary_avg.loc[scenario, sac_policy_name]
        
        # 在寻找最佳基准时，排除所有版本的SAC
        sac_variants_to_drop = ['RL (SAC) 无约束', 'RL (SAC) 带1%约束', 'RL (SAC) 动态约束']
        baselines_incomes = summary_avg.loc[scenario].drop(sac_variants_to_drop)
        # ====================================================================

        best_baseline_income = baselines_incomes.max()
        best_baseline_name = baselines_incomes.idxmax()
        improvement_pct = (sac_income / best_baseline_income - 1) * 100
        
        print(f"在【{scenario}】场景下:")
        print(f"  - {sac_policy_name} 平均收入: {sac_income:.2f} 万元")
        print(f"  - 最佳基准策略 ({best_baseline_name}): {best_baseline_income:.2f} 万元")
        print(f"  - SAC 性能提升百分比: {improvement_pct:.2f}%\n")

    # (后续的最差情况表现和夏普比率计算，无需修改，会自动包含新策略)
    summary_min = df.groupby(['scenario', 'policy'])['total_income'].min().unstack()
    print("\n--- 风险指标：最差情况表现 (万元) ---\n数值越大，抗风险能力越强：")
    print(summary_min.round(2))

    df['sharpe_ratio'] = df['daily_incomes'].apply(calculate_sharpe)
    summary_sharpe = df.groupby(['scenario', 'policy'])['sharpe_ratio'].mean().unstack()
    print("\n--- 风险指标：平均夏普比率 ---\n数值越大，风险调整后收益越高：")
    print(summary_sharpe.round(2))


# --- [ 步骤 4: 自动化决策行为诊断分析 (无需修改) ] ---
# (这部分代码是健壮的，会自动包含并分析所有策略，包括新的动态约束SAC)
if not df.empty:
    case_study_runs = df.dropna(subset=['case_study_movie'])
    unique_cases = case_study_runs.drop_duplicates(subset=['start_date', 'case_study_movie'])

    plots_dir = '../plots'
    os.makedirs(plots_dir, exist_ok=True)
    all_speed_results = []

    print("\n" + "="*80)
    print("### 决策行为诊断指标分析 ###")
    print("="*80)
    
    for index, case_info in unique_cases.iterrows():
        case_name = case_info['case_study_movie']
        case_date = case_info['start_date']
        
        if case_name in ['流浪地球', '这个杀手不太冷静', '熊出没·伴我“熊芯”']:
            case_type = '黑马'; metric_func = calculate_capture_speed; threshold = 0.20; unit = '捕获速度 (天)'
        else:
            case_type = '高开低走'; metric_func = calculate_cut_loss_speed; threshold = 0.05; unit = '止损速度 (天)'

        print(f"\n--- 案例分析：{case_type} - 《{case_name}》 ---")
        
        case_df = df[df['start_date'] == case_date].copy()
        
        metric_col_name = unit.split(' ')[0]
        case_df[metric_col_name] = case_df.apply(
            lambda row: metric_func(row['daily_actions'], row['case_study_index'], threshold), axis=1
        )
        speed_summary = case_df.groupby('policy')[metric_col_name].mean().sort_values()
        speed_summary.name = f"《{case_name}》 {unit}"
        all_speed_results.append(speed_summary)
        
        print(f"速度对比 (数值越小越好):")
        print(speed_summary)
        
        plt.figure(figsize=(12, 7))
        for policy_name in speed_summary.index:
            run_data = case_df[case_df['policy'] == policy_name].iloc[0]
            movie_idx = run_data['case_study_index']
            if not np.isnan(movie_idx):
                movie_idx = int(movie_idx)
                if len(run_data['daily_actions'][0]) > movie_idx:
                    schedule_over_time = [s[movie_idx] * 100 for s in run_data['daily_actions']]
                    plt.plot(range(1, len(schedule_over_time) + 1), schedule_over_time, marker='.', linestyle='-', label=policy_name)
        
        plt.axhline(y=threshold * 100, color='r', linestyle='--', label=f"阈值 ({threshold*100:.0f}%)")
        plt.title(f"各策略对“{case_type}”《{case_name}》的排片变化曲线", fontsize=16)
        plt.xlabel('模拟天数 (Day)', fontsize=12); plt.ylabel('排片占比 (%)', fontsize=12)
        plt.legend(); plt.grid(True, which='both', linestyle='--', linewidth=0.5)
        
        file_name = f"{case_name}_{metric_col_name}.png"
        save_path = os.path.join(plots_dir, file_name)
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
        print(f"✅ 图表已保存至: {save_path}")
        plt.show()

    if all_speed_results:
        final_summary_df = pd.concat(all_speed_results, axis=1)
        print("\n" + "="*80)
        print("### 所有案例诊断指标汇总表 ###")
        print("数值代表天数，越小表示策略反应越快、越优秀")
        print(final_summary_df.round(2))
        print("="*80)
```
