```python
# -*- coding: utf-8 -*-
import pandas as pd
import numpy as np
import os
import json
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.colors as mcolors
import sys
import platform

# ==============================================================================
# 0. ÂÖ®Â±ÄÈÖçÁΩÆÔºöÁªòÂõæÈ£éÊ†º‰∏éÈÖçËâ≤ (Global Config)
# ==============================================================================
def set_pub_style():
    sys_name = platform.system()
    if sys_name == "Windows":
        font_list = ['Microsoft YaHei', 'SimHei', 'Arial', 'sans-serif']
    elif sys_name == "Darwin":
        font_list = ['PingFang SC', 'Heiti TC', 'Arial Unicode MS', 'Arial', 'sans-serif']
    else:
        font_list = ['WenQuanYi Micro Hei', 'Noto Sans CJK SC', 'SimHei', 'Arial', 'sans-serif']
        
    plt.rcParams['font.family'] = ['sans-serif']
    plt.rcParams['font.sans-serif'] = font_list
    plt.rcParams['axes.unicode_minus'] = False
    plt.rcParams['axes.titlesize'] = 15
    plt.rcParams['axes.labelsize'] = 13
    plt.rcParams['xtick.labelsize'] = 11
    plt.rcParams['ytick.labelsize'] = 11
    plt.rcParams['legend.fontsize'] = 10
    plt.rcParams['axes.linewidth'] = 1.2
    plt.rcParams['lines.linewidth'] = 2.0
    plt.rcParams['figure.dpi'] = 150

# --- ÁªèÂÖ∏È´òÂØπÊØîÈÖçËâ≤ (AAAS Style) ---
COLOR_PALETTE = {
    "SAC": "#0C5DA5",        # ÁßëÂ≠¶Ëìù
    "Efficiency": "#00B945", # È≤úÁªø
    "Greedy": "#FF9500",     # ‰∫ÆÊ©ô
    "Static": "#FF2C00",     # Á∫ØÁ∫¢
    "Ref": "#7F7F7F"         # ÁÅ∞Ëâ≤
}

# ==============================================================================
# 1. Êï∞ÊçÆÂä†ËΩΩ‰∏éÂêçÂçï
# ==============================================================================
def load_data(csv_path='logs/main4/main_evaluation_results.csv'):
    try:
        current_dir = os.path.dirname(os.path.abspath(__file__))
    except NameError:
        current_dir = os.getcwd()

    possible_paths = [
        csv_path, 
        os.path.join(current_dir, csv_path),
        os.path.join(current_dir, 'logs', 'main4', 'main_evaluation_results.csv'),
        os.path.join(current_dir, 'logs', 'evaluation_all', 'main_evaluation_results.csv'),
        os.path.join(os.path.dirname(current_dir), 'logs','main4', 'main_evaluation_results.csv')
    ]
    
    found_path = None
    for p in possible_paths:
        if os.path.exists(p):
            found_path = p
            break
            
    if not found_path:
        print(f"‚ùå ÈîôËØØÔºöÂú®‰ª•‰∏ãË∑ØÂæÑÂùáÊú™ÊâæÂà∞Êï∞ÊçÆÊñá‰ª∂:\n{possible_paths}")
        return pd.DataFrame()

    print(f"üìÇ Ê≠£Âú®ËØªÂèñÊï∞ÊçÆ: {found_path}")
    try:
        df = pd.read_csv(found_path, encoding='utf-8')
    except UnicodeDecodeError:
        try:
            df = pd.read_csv(found_path, encoding='gbk')
        except Exception as e:
            print(f"‚ùå ËØªÂèñÂ§±Ë¥•: {e}")
            return pd.DataFrame()

    for col in ['daily_actions', 'daily_incomes']:
        if col in df.columns:
            df[col] = df[col].apply(lambda x: json.loads(x) if isinstance(x, str) else x)
    
    if 'case_study_index' in df.columns:
        df['case_study_index'] = pd.to_numeric(df['case_study_index'], errors='coerce')

    return df

DARK_HORSE_MOVIES = [
    'ÊµÅÊµ™Âú∞ÁêÉ', 'ÁÜäÂá∫Ê≤°¬∑ÂéüÂßãÊó∂‰ª£', 'Êñ∞ÂñúÂâß‰πãÁéã',
    'Âà∫ÊùÄÂ∞èËØ¥ÂÆ∂', '‰∫∫ÊΩÆÊ±πÊ∂å', 'ÁÜäÂá∫Ê≤°¬∑ÁãÇÈáéÂ§ßÈôÜ',
    'Ëøô‰∏™ÊùÄÊâã‰∏çÂ§™ÂÜ∑Èùô', 'ÁÜäÂá∫Ê≤°¬∑ÈáçËøîÂú∞ÁêÉ', 'ÁãôÂáªÊâã', 'ÂõõÊµ∑',
    'ÁÜäÂá∫Ê≤°¬∑‰º¥Êàë‚ÄúÁÜäËäØ‚Äù', 'Êó†Âêç', 'Ê∑±Êµ∑', 'Á¨¨‰∫åÂçÅÊù°', 'ÁÜäÂá∫Ê≤°¬∑ÈÄÜËΩ¨Êó∂Á©∫',
    'ÊàëÂíåÊàëÁöÑÁà∂Ëæà', 'ÂùöÂ¶ÇÁ£êÁü≥', 'Ââç‰ªª4ÔºöËã±Âπ¥Êó©Â©ö', 'ÂøóÊÑøÂÜõÔºöÈõÑÂÖµÂá∫Âáª', 'Ëé´ÊñØÁßëË°åÂä®',
    'ÊÇ¨Â¥ñ‰πã‰∏ä', '‰∫∫ÁîüË∑Ø‰∏çÁÜü', 'Êú´Ë∑ØÁãÇËä±Èí±', '‰πùÈæôÂüéÂØ®‰πãÂõ¥Âüé',
    'Â∞ÅÁ•ûÁ¨¨‰∏ÄÈÉ®ÔºöÊúùÊ≠åÈ£é‰∫ë', 'ÈïøÂÆâ‰∏â‰∏áÈáå', 'Â≠¶Áà∏', 'ÂºÇÂΩ¢ÔºöÂ§∫ÂëΩËà∞',
    'ÊØîÊÇ≤‰º§Êõ¥ÊÇ≤‰º§ÁöÑÊïÖ‰∫ã', 'ÊàëÁöÑÂßêÂßê', 'Êâ¨ÂêçÁ´ã‰∏á', 'ÈáëÊâãÊåá', 'Âë®Â§ÑÈô§‰∏âÂÆ≥', '‰Ω†ÊÉ≥Ê¥ªÂá∫ÊÄéÊ†∑ÁöÑ‰∫∫Áîü'
]

ROTTEN_MOVIES = [
    'ÂìÜÂï¶AÊ¢¶ÔºöÂ§ßÈõÑÁöÑÊúàÁêÉÊé¢Èô©ËÆ∞', '‰∏äÊµ∑Â†°Âûí', 'Â∞èQ',
    'Â§öÂäõÁâπÁöÑÂ•áÂπªÂÜíÈô©', 'ÂñãË°ÄÊàòÂ£´', 'ÊäµËææ‰πãË∞ú', '1917', 'Êï∞Á†ÅÂÆùË¥ùÂ§ßÂÜíÈô©ÔºöÊúÄÂêéÁöÑËøõÂåñ', 
    'Âú∞Áã±Áî∑ÁàµÔºöË°ÄÁöáÂêéÂ¥õËµ∑', '‰∏ÄÁßíÈíü',
    'Áå´ÂíåËÄÅÈº†', '‰Ω†Â•Ω‰∏ñÁïå', '‰∫Ü‰∏çËµ∑ÁöÑËÄÅÁà∏', 'ÊàëÊ≤°Ë∞àÂÆåÁöÑÈÇ£Âú∫ÊÅãÁà±', 'Èù©ÂëΩËÄÖ', 'Ê¢ÖËâ≥Ëä≥', 'ÈìÅÈÅìËã±ÈõÑ',
    'ÈªëÂÆ¢Â∏ùÂõΩÔºöÁü©ÈòµÈáçÂêØ', 'Á•ûÁßòÊµ∑Âüü', 'Á≤æÁÅµÊóÖÁ§æ4ÔºöÂèòË∫´Â§ßÂÜíÈô©', 'ËæπÁºòË°åËÄÖ', '‰∫°ÂëΩÊïëÊä§ËΩ¶', 
    'ÂìÜÂï¶AÊ¢¶ÔºöÂ§ßÈõÑÁöÑÂÆáÂÆôÂ∞èÊàò‰∫â2021', '‰∏ÄÂë®ÁöÑÊúãÂèã',
    '‰∏çËÉΩÊµÅÊ≥™ÁöÑÊÇ≤‰º§', 'ÂìÜÂï¶AÊ¢¶ÔºöÂ§ßÈõÑ‰∏éÂ§©Á©∫ÁöÑÁêÜÊÉ≥‰π°', 'Â§©Á©∫‰πãÂüé', 'ÂøµÂøµÁõ∏Âøò',
    'ÂìÜÂï¶AÊ¢¶ÔºöÂ§ßÈõÑÁöÑÂú∞ÁêÉ‰∫§Âìç‰πê', 'ÊéíÁêÉÂ∞ëÂπ¥!! ÂûÉÂúæÂú∫ÂÜ≥Êàò', 'ÂºÇ‰∫∫‰πã‰∏ã'
]

# ==============================================================================
# 2. Èò∂ÊÆµ‰∏ÄÔºöÂçï‰ΩìÂàÜÊûê (Individual Analysis)
# ==============================================================================
def run_individual_analysis(df, plots_dir):
    print(f"\nüöÄ [Èò∂ÊÆµ‰∏Ä] ÂºÄÂßãÁªòÂà∂Âçï‰ΩìÊ°à‰æãÂõæË°®...")
    
    unique_cases = df[['start_date', 'case_study_movie']].drop_duplicates()
    
    PLOT_DAYS = 7      
    PENALTY_DAYS = 7    
    
    stats_data = {"Black Horse": {}, "Rotten/Flop": {}}
    detailed_records = []

    for _, row in unique_cases.iterrows():
        date = row['start_date']
        movie = row['case_study_movie']
        
        if movie in DARK_HORSE_MOVIES:
            category = "Black Horse"
            threshold_val = 35.0
            threshold_label = "Capture (35%)"
            title_suffix = "(Black Horse)"
            legend_loc = "upper left"
        elif movie in ROTTEN_MOVIES:
            category = "Rotten/Flop"
            threshold_val = 10.0
            threshold_label = "Stop Loss (10%)"
            title_suffix = "(High Start, Low Finish)"
            legend_loc = "upper right"
        else:
            continue

        case_df = df[(df['start_date'] == date) & (df['case_study_movie'] == movie)].copy()
        if case_df.empty: continue
            
        policy_map = {
            "RL (SAC)": {"label": "Ours (SAC)", "color": COLOR_PALETTE["SAC"], "z": 10},
            "ÊïàÁéáÂêØÂèëÂºèÁ≠ñÁï•": {"label": "Efficiency", "color": COLOR_PALETTE["Efficiency"], "z": 3},
            "Ë¥™Â©™ÂêØÂèëÂºèÁ≠ñÁï•": {"label": "Greedy", "color": COLOR_PALETTE["Greedy"], "z": 2},
            "ÈùôÊÄÅÂêØÂèëÂºèÁ≠ñÁï•": {"label": "Static", "color": COLOR_PALETTE["Static"], "z": 1} 
        }
        
        draw_order = ["RL (SAC)", "ÊïàÁéáÂêØÂèëÂºèÁ≠ñÁï•", "Ë¥™Â©™ÂêØÂèëÂºèÁ≠ñÁï•", "ÈùôÊÄÅÂêØÂèëÂºèÁ≠ñÁï•"]
        existing_policies = case_df['policy'].unique()

        for policy_name_cn in draw_order:
            if policy_name_cn not in existing_policies: continue 
            if policy_name_cn not in policy_map: continue

            run = case_df[case_df['policy'] == policy_name_cn].iloc[0]
            idx = run['case_study_index']
            if pd.isna(idx): continue
            idx = int(idx)
            
            actions = run['daily_actions']
            y_values = []
            for day_act in actions[:PLOT_DAYS]:
                if len(day_act) > idx:
                    y_values.append(day_act[idx] * 100) 
                else:
                    y_values.append(0)
            
            style = policy_map[policy_name_cn]
            
            # Êï∞ÊçÆÁªüËÆ°
            captured_day = None
            simple_label = style["label"]
            
            if simple_label not in stats_data[category]:
                stats_data[category][simple_label] = []

            for day_idx, share in enumerate(y_values):
                current_day = day_idx + 1
                is_captured = False
                if category == "Black Horse":
                    if share >= 35.0: is_captured = True
                elif category == "Rotten/Flop":
                    if share <= 10.0: is_captured = True
                
                if is_captured:
                    captured_day = current_day
                    break
            
            final_day = captured_day if captured_day is not None else PENALTY_DAYS
            stats_data[category][simple_label].append(final_day)
            detailed_records.append({
                "Category": category,
                "Movie": movie,
                "Strategy": simple_label,
                "Day": final_day
            })

    print(f"‚úÖ [Èò∂ÊÆµ‰∏ÄÂÆåÊàê] ÊâÄÊúâÂçï‰ΩìÊ°à‰æãÊï∞ÊçÆÁªüËÆ°ÂÆåÊØï„ÄÇ")
    return stats_data, detailed_records, PENALTY_DAYS

# ==============================================================================
# 3. Èò∂ÊÆµ‰∫åÔºöÊ±áÊÄªÂàÜÊûê (ÂçïÂùêÊ†áËΩ¥Ôºå‰ΩçÁΩÆÂàÜÁªÑ)
# ==============================================================================

def _decorate_single_axis(ax, penalty_days, group_labels, group_centers):
    """
    Áªü‰∏ÄË£ÖÈ•∞ÂùêÊ†áËΩ¥ÔºåÊ∑ªÂä†ÂàÜÁªÑÊ†áÁ≠æ
    """
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.2)
    ax.spines['bottom'].set_linewidth(1.2)
    
    ax.set_ylabel("Capture Time (Days)", fontsize=12, fontweight='bold')
    ax.yaxis.set_major_locator(ticker.MaxNLocator(integer=True))
    ax.set_ylim(bottom=0, top=penalty_days + 1.5)
    ax.grid(True, axis='y', linestyle='--', color='#E0E0E0', alpha=0.8, zorder=0)

    # Âú®È°∂ÈÉ®Ê∑ªÂä†ÂàÜÁªÑÊ†áÁ≠æ (Black Horse vs Rotten)
    y_max = ax.get_ylim()[1]
    for label, center in zip(group_labels, group_centers):
        # Áîª‰∏ÄÊù°Áü≠Ê®™Á∫øË°®Á§∫ÂàÜÁªÑËåÉÂõ¥ (ÂèØÈÄâ)
        # ax.plot([center-1.5, center+1.5], [y_max, y_max], color='black', lw=1.5, clip_on=False)
        ax.text(center, y_max * 0.95, label, ha='center', va='bottom', fontsize=13, fontweight='bold')

def run_summary_analysis(stats_data, detailed_records, penalty_days, plots_dir):
    print(f"\nüöÄ [Èò∂ÊÆµ‰∫å] ÂºÄÂßãÊ±áÊÄªÁªüËÆ°ÂàÜÊûê (ÂçïÂùêÊ†áËΩ¥ÂàÜÁªÑÁªòÂà∂)...")

    policy_order = ["Ours (SAC)", "Efficiency", "Greedy", "Static"]
    # ÁõÆÊ†áÈ°∫Â∫èÔºöÂÖà Black HorseÔºåÂêé Rotten/Flop
    target_categories = ["Black Horse", "Rotten/Flop"]
    
    # ----------------------------------------------------
    # 1. ÂáÜÂ§áÊï∞ÊçÆÁªìÊûÑÔºöÂ∞Ü‰∏§ÁªÑÊï∞ÊçÆÂêàÂπ∂Âà∞‰∏Ä‰∏™ÂàóË°®‰∏≠ÔºåÂπ∂ËÆ°ÁÆó‰ΩçÁΩÆ
    # ----------------------------------------------------
    combined_data = []
    combined_colors = []
    combined_labels = []
    combined_positions = []
    
    # ÂÆö‰πâ‰ΩçÁΩÆÂèÇÊï∞
    # Á¨¨‰∏ÄÁªÑ(Black Horse): 1, 2, 3, 4
    # Èó¥Ë∑ù (Gap): 2 (Á©∫Âá∫ 5, 6)
    # Á¨¨‰∫åÁªÑ(Rotten): 7, 8, 9, 10
    
    current_pos_start = 1
    pos_step = 1.0
    group_gap = 1.0  # ‰∏§‰∏™Â§ßÁ±ª‰πãÈó¥ÁöÑË∑ùÁ¶ª
    
    group_center_positions = [] # Áî®‰∫éÊ†áËÆ∞Â§ßÁ±ªÂêçÁß∞ÁöÑ‰ΩçÁΩÆ
    
    for category in target_categories:
        strategies = stats_data.get(category, {})
        
        # ‰∏¥Êó∂Â≠òÂÇ®Ëøô‰∏ÄÁªÑÁöÑ‰ΩçÁΩÆÔºåÁî®‰∫éËÆ°ÁÆó‰∏≠ÂøÉÁÇπ
        current_group_indices = []
        
        for i, policy in enumerate(policy_order):
            # ËÆ°ÁÆóÂΩìÂâçÁÆ±Â≠êÁöÑÁªùÂØπ‰ΩçÁΩÆ
            pos = current_pos_start + i * pos_step
            current_group_indices.append(pos)
            
            # Ëé∑ÂèñÊï∞ÊçÆ
            if strategies and policy in strategies and strategies[policy]:
                combined_data.append(strategies[policy])
            else:
                combined_data.append([]) # Á©∫Êï∞ÊçÆÂç†‰Ωç
                
            # Ëé∑ÂèñÊ†áÁ≠æÂíåÈ¢úËâ≤
            combined_labels.append(policy)
            if "SAC" in policy: key = "SAC"
            elif "Efficiency" in policy: key = "Efficiency"
            elif "Greedy" in policy: key = "Greedy"
            elif "Static" in policy: key = "Static"
            else: key = "Ref"
            combined_colors.append(COLOR_PALETTE[key])
            
            combined_positions.append(pos)
            
        # ËÆ∞ÂΩïËØ•ÁªÑÁöÑ‰∏≠ÂøÉ‰ΩçÁΩÆÔºåÊñπ‰æøÂÜôÊ†áÈ¢ò
        if current_group_indices:
            center = sum(current_group_indices) / len(current_group_indices)
            group_center_positions.append(center)
        
        # Êõ¥Êñ∞‰∏ã‰∏ÄÁªÑÁöÑËµ∑Âßã‰ΩçÁΩÆ = ÂΩìÂâçÁªÑÊúÄÂêé‰∏Ä‰∏™‰ΩçÁΩÆ + ÁªÑÈó¥Ë∑ù + 1 (Âõ†‰∏∫ i ‰ªé 0 ÂºÄÂßã)
        current_pos_start = current_group_indices[-1] + group_gap + pos_step

    # Â¶ÇÊûúÊ≤°Êúâ‰ªª‰ΩïÊï∞ÊçÆÔºåÁõ¥Êé•ÈÄÄÂá∫
    if not any(len(d) > 0 for d in combined_data):
        print("Ê≤°ÊúâÊúâÊïàÁªòÂõæÊï∞ÊçÆ„ÄÇ")
        return

    # ==================================================
    # ÁªòÂà∂Âõæ 1: ÂçïËΩ¥ÂàÜÁªÑÁÆ±Á∫øÂõæ (Single Axis Box Plot)
    # ==================================================
    fig_box, ax_box = plt.subplots(figsize=(12, 6))
    
    # 1. ÁªòÂà∂ÁÆ±Á∫øÂõæ (‰ΩøÁî® positions ÂèÇÊï∞ÊéßÂà∂‰ΩçÁΩÆ)
    bp = ax_box.boxplot(combined_data, positions=combined_positions, 
                        patch_artist=True, showfliers=False, widths=0.6, zorder=1)
    
    # 2. ËÆæÁΩÆÊ†∑Âºè
    for patch, color in zip(bp['boxes'], combined_colors):
        rgba_color = mcolors.to_rgba(color, alpha=0.3)
        patch.set_facecolor(rgba_color)
        patch.set_edgecolor(color)
        patch.set_linewidth(1.5)
        
    for element in ['whiskers', 'caps']:
        plt.setp(bp[element], color='black', linewidth=1.2, alpha=0.8)
    plt.setp(bp['medians'], color='#D32F2F', linewidth=2.0)

    # 3. ÁªòÂà∂Êï£ÁÇπ
    for i, d in enumerate(combined_data):
        # Âú®ÂØπÂ∫îÁöÑ position ‰∏äÂä† jitter
        pos = combined_positions[i]
        x_jitter = np.random.normal(pos, 0.05, size=len(d))
        ax_box.scatter(x_jitter, d, alpha=0.8, color=combined_colors[i], s=30, 
                       edgecolor='white', linewidth=0.6, zorder=2)
    
    # 4. ËÆæÁΩÆ X ËΩ¥Ê†áÁ≠æ
    ax_box.set_xticks(combined_positions)
    # ÁÆÄÂåñÊ†áÁ≠æÔºöÂ¶ÇÊûúÊòØÈáçÂ§çÁöÑÁ≠ñÁï•ÂêçÔºåÂÖ∂ÂÆûÂèØ‰ª•Âè™Âú®Âõæ‰æãÊòæÁ§∫ÔºåËøôÈáå‰∏∫‰∫ÜÊ∏ÖÊô∞ËøòÊòØ‰øùÁïôXËΩ¥ÊñáÂ≠ó
    # ‰πüÂèØ‰ª•ËÄÉËôëÊóãËΩ¨Â≠ó‰ΩìÈò≤Ê≠¢ÈáçÂè†
    ax_box.set_xticklabels(combined_labels, rotation=0, fontsize=10)
    
    # 5. Ë£ÖÈ•∞‰∏éÂàÜÁªÑÊ†áËÆ∞
    _decorate_single_axis(ax_box, penalty_days, target_categories, group_center_positions)
    ax_box.set_title("Capture Time Distribution (Box Plot)", fontsize=16, fontweight='bold', pad=20)
    
    # ‰øùÂ≠ò
    filename_box = "Summary_SingleAxis_BoxPlot.png"
    save_path_box = os.path.join(plots_dir, filename_box)
    fig_box.savefig(save_path_box, dpi=300, bbox_inches='tight', pad_inches=0.1)
    print(f"   -> Â∑≤‰øùÂ≠òÂçïËΩ¥ÂàÜÁªÑÁÆ±Á∫øÂõæ: {filename_box}")
    plt.close(fig_box)


    # ==================================================
    # ÁªòÂà∂Âõæ 2: ÂçïËΩ¥ÂàÜÁªÑÂ∞èÊèêÁê¥Âõæ (Single Axis Violin Plot)
    # ==================================================
    fig_violin, ax_violin = plt.subplots(figsize=(12, 6))
    
    # 1. ÁªòÂà∂Â∞èÊèêÁê¥ (‰ΩøÁî® positions ÂèÇÊï∞)
    vparts = ax_violin.violinplot(combined_data, positions=combined_positions, widths=0.8, 
                                  showmeans=False, showmedians=False, showextrema=False)
    
    for pc, color in zip(vparts['bodies'], combined_colors):
        pc.set_facecolor(color)
        pc.set_edgecolor(color)
        pc.set_alpha(0.4)
        
    # 2. ÁªòÂà∂Êï£ÁÇπ
    for i, d in enumerate(combined_data):
        pos = combined_positions[i]
        x_jitter = np.random.normal(pos, 0.05, size=len(d))
        ax_violin.scatter(x_jitter, d, alpha=0.8, color=combined_colors[i], s=30, 
                          edgecolor='white', linewidth=0.6, zorder=2)

    # 3. ËÆæÁΩÆ X ËΩ¥
    ax_violin.set_xticks(combined_positions)
    ax_violin.set_xticklabels(combined_labels, rotation=0, fontsize=10)
    
    # 4. Ë£ÖÈ•∞
    _decorate_single_axis(ax_violin, penalty_days, target_categories, group_center_positions)
    ax_violin.set_title("Capture Time Distribution (Violin Plot)", fontsize=16, fontweight='bold', pad=20)

    # ‰øùÂ≠ò
    filename_violin = "Summary_SingleAxis_ViolinPlot.png"
    save_path_violin = os.path.join(plots_dir, filename_violin)
    fig_violin.savefig(save_path_violin, dpi=300, bbox_inches='tight', pad_inches=0.1)
    print(f"   -> Â∑≤‰øùÂ≠òÂçïËΩ¥ÂàÜÁªÑÂ∞èÊèêÁê¥Âõæ: {filename_violin}")
    plt.close(fig_violin)


    # --- ËæìÂá∫Ë°®Ê†º (‰øùÊåÅ‰∏çÂèò) ---
    print("\n" + "="*80)
    print("üìã ÁîµÂΩ±ÊçïËé∑Êó∂Èó¥ËØ¶Ë°®")
    print("="*80)
    if detailed_records:
        df_detailed = pd.DataFrame(detailed_records)
        pivot_table = df_detailed.pivot_table(index=['Category', 'Movie'], columns='Strategy', values='Day', aggfunc='first')
        desired_order = ["Ours (SAC)", "Efficiency", "Greedy", "Static"]
        existing_cols = [c for c in desired_order if c in pivot_table.columns]
        pivot_table = pivot_table[existing_cols]
        pd.set_option('display.max_rows', None)
        print(pivot_table)

    print("\n" + "="*60)
    print(f"üìä Á≠ñÁï•Âπ≥ÂùáÊçïËé∑Êó∂Èó¥Ê±áÊÄª (Êú™ÊçïËé∑ËÆ∞‰∏∫ {penalty_days})")
    print("="*60)
    for cat_name, strategies in stats_data.items():
        print(f"\nüìå Category: {cat_name}")
        policy_order_print = ["Ours (SAC)", "Efficiency", "Greedy", "Static"]
        for strat_name in policy_order_print:
            if strat_name in strategies:
                times = strategies[strat_name]
                if times:
                    avg_time = sum(times) / len(times)
                    print(f"   - {strat_name:<12}: {avg_time:.2f}")
    print("="*60)
    print(f"‚úÖ [Èò∂ÊÆµ‰∫åÂÆåÊàê] ÊâÄÊúâÁªüËÆ°ÂàÜÊûêÂ∑≤ÁªìÊùü„ÄÇ")

# ==============================================================================
# 4. ‰∏ªÁ®ãÂ∫èÂÖ•Âè£
# ==============================================================================
def main():
    set_pub_style()
    df = load_data()
    if df.empty: return

    try:
        current_dir = os.path.dirname(os.path.abspath(__file__))
    except NameError:
        current_dir = os.getcwd()

    plots_dir = os.path.join(current_dir, 'plots', 'main4_analysis')
    os.makedirs(plots_dir, exist_ok=True)
    print(f"üìÇ ÂõæÁâáÂ∞Ü‰øùÂ≠òËá≥: {plots_dir}")

    # 1. ËøêË°åÂçï‰ΩìÂàÜÊûê
    stats_data, detailed_records, penalty_days = run_individual_analysis(df, plots_dir)
    
    # 2. ËøêË°åÊ±áÊÄªÂàÜÊûê (ÂçïËΩ¥ÂàÜÁªÑÁªòÂà∂)
    run_summary_analysis(stats_data, detailed_records, penalty_days, plots_dir)

if __name__ == "__main__":
    main()
```
